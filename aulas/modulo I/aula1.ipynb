## Exercício 4: Média de Notas

```python
print( "Media de Notas" )
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))
media = (nota1 + nota2 + nota3) / 3

print(f'A sua primeira nota foi {nota1}, a segunda nota foi {nota2} e a terceira nota foi {nota3}.')
print(f"A média das notas é: {round(media, 2)}")
```

## Exercício 5: Cálculo de Horas Extras

Objetivo: calcular o salário total considerando horas normais e horas extras (com acréscimo de 50%).

```python
print( "Cálculo de horas extras" )

nome_completo = input("Digite o seu nome completo: ")
valor_hora_normal = float(input("Digite o valor da hora normal: "))
horas_trabalhadas_normais = float(input("Digite o número de horas normais trabalhadas: "))
horas_extras_trabalhadas = float(input("Digite o número de horas extras trabalhadas: "))
valor_hora_extra = valor_hora_normal * 1.5

salario_total = (valor_hora_normal * horas_trabalhadas_normais) + (valor_hora_extra * horas_extras_trabalhadas)

print(f"O {nome_completo}, o seu salário total é: {salario_total}")
```

## Exercício 6: Conversor de Distâncias

Objetivo: converter distâncias entre km, metros e centímetros usando condicionais (`if/elif/else`).

```python

distancia = float(input("Digite a distância percorrida: "))

unidade = input("Digite a unidade da distância (km, m, cm): ")

unidade_convertida = input("Digite a unidade para a qual deseja converter (km, m, cm): ")

if unidade == "km":
    distancia_metros = distancia * 1000
elif unidade == "m":
    distancia_metros = distancia
elif unidade == "cm":
    distancia_metros = distancia / 100
else:
    print("Unidade de distância inválida.")
    exit()
    
if unidade_convertida == "km":
    distancia_convertida = distancia_metros / 1000
elif unidade_convertida == "m":
    distancia_convertida = distancia_metros
elif unidade_convertida == "cm":
    distancia_convertida = distancia_metros * 100
else:
    print("Unidade de conversão inválida.")
    exit()
print(f"A distância convertida é: {distancia_convertida} {unidade_convertida}")

```

## Exercício 8: Calculadora de IMC

Objetivo: calcular o Índice de Massa Corporal e classificar o resultado usando condicionais.

| IMC | Classificação |
|-----|--------------|
| < 18.5 | Abaixo do peso |
| 18.5 - 24.9 | Peso normal |
| 25.0 - 29.9 | Sobrepeso |
| >= 30.0 | Obesidade |

```python
altura = float(input("Digite a sua altura em metros: "))
peso = float(input("Digite o seu peso em kg: "))

imc = peso / (altura ** 2)

if imc < 18.5:
    classificacao = "Abaixo do peso" 
elif imc < 25:
    classificacao = "Peso normal"
elif imc < 30:
    classificacao = "Sobrepeso"
else:
    classificacao = "Obesidade"

print(f"O seu IMC é: {imc:.2f}")
print(f"A sua classificação de IMC é: {classificacao}")

```

## Exercício 9: Simulador de Investimento

Objetivo: calcular o valor futuro de um investimento com juros compostos e aportes mensais.

```python
valor_investido = float(input("Digite o valor investido: "))
tempo_investimento = float(input("Digite o tempo do investimento em anos: "))
taxa_juros =  float(input("Digite a taxa de juros anual (em %): "))
valor_adicional_mensal = float(input("Digite o valor adicional mensal (se houver, caso contrário digite 0): "))

valor_final = valor_investido * (1 + (taxa_juros / 100)) ** tempo_investimento

if valor_adicional_mensal > 0:
    valor_final += valor_adicional_mensal * tempo_investimento * 12


print(f"O valor final do investimento após {tempo_investimento} anos é: R$ {valor_final:.2f}")
```

## Exercício 10: Gerador de Recibo

Objetivo: montar um recibo de pagamento formatado com dados do cliente, serviço e valores.

```python
nome_cliente = input("Digite o nome do cliente: ")
servico_contratado = input("Digite o serviço contratado: ")
valor_servico = float(input("Digite o valor do serviço: "))
horas_trabalhadas = float(input("Digite o número de horas trabalhadas: "))
valor_hora  = float(input("Digite o valor da hora trabalhada: "))
total = valor_servico + (horas_trabalhadas * valor_hora)

print("===== Recibo de Pagamento =====")
print(f"Cliente: {nome_cliente}")
print(f"Serviço: {servico_contratado}")
print(f"Valor do serviço: R$ {valor_servico:.2f}")
print(f"Horas trabalhadas: {horas_trabalhadas}")
print(f"Valor da hora: R$ {valor_hora:.2f}")
print(f"Total: R$ {total:.2f}")
print("==============================")
```
