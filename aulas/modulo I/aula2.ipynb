## TypeError

Um TypeError ocorre em Python quando uma operação ou função é aplicada a um objeto de tipo inadequado. Python não sabe como aplicar a operação porque os tipos de dados não são compatíveis.

## Exemplo de TypeError
Um exemplo clássico é tentar utilizar a função len() com um inteiro, o que resulta em TypeError, pois len() espera um objeto iterável, como uma string, lista, ou tupla, não um inteiro.

# Aula 02 | TypeError, Type Check e Tratamento de Erros

Nesta aula você vai aprender a lidar com erros em Python:
- **TypeError** — quando tipos de dados são incompatíveis
- **Type Check** — como verificar o tipo de uma variável
- **try/except** — tratamento de exceções para código robusto
- **Validação de entrada** — garantir que o usuário forneça dados válidos

Saber tratar erros é essencial na engenharia de dados, onde os dados de entrada nem sempre vêm no formato esperado.

```python
# Exemplo que causa TypeError
try:
    resultado = len(5)
except TypeError as e:
    print(e)  # Imprime a mensagem de erro
```

O código acima tenta obter o comprimento de um inteiro, o que não faz sentido, resultando na mensagem de erro: "object of type 'int' has no len()".

## Type Check

Verificação de tipo (type check) é o processo de verificar o tipo de uma variável. Isso pode ser útil para garantir que operações ou funções sejam aplicadas apenas a tipos de dados compatíveis, evitando erros em tempo de execução.

### Exemplo de Type Check
Para verificar o tipo de uma variável em Python, você pode usar a função type() ou isinstance().

```python
numero = 10
if isinstance(numero, int):
    print("A variável é um inteiro.")
else:
    print("A variável não é um inteiro.")
```

Este código verifica se numero é uma instância de int e imprime uma mensagem apropriada.

Agora vamos aplicar os conceitos de tratamento de erros em programas interativos.

### Exercício 1: Conversor de Temperatura com Validação

Objetivo: converter Celsius para Fahrenheit tratando entradas vazias e inválidas com `try/except`.

```python
print("Conversor de Temperatura")

try:
    celsius_info = input("Informe a temperatura em Celsius: ").strip()

    if celsius_info == "":
        raise ValueError("Erro: entrada vazia.")

    celsius = float(celsius_info)
    fahrenheit = (celsius * 9/5) + 32
    print(f"A temperatura em Fahrenheit é {fahrenheit:.2f}")

except ValueError as e:
    print(e)
```

### Exercício 2: Verificador de Palíndromo

Objetivo: verificar se uma palavra ou frase é um palíndromo, removendo espaços e caracteres especiais.

```python
print("Verificador de Palíndromo")

try:
    entrada = input("Insira uma entrada para verificarmos se é um palíndromo: ").strip().lower()

    if entrada == "":
        raise ValueError("Erro: entrada vazia")

    texto_limpo = ""

    for caractere in entrada:
        if caractere.isalnum():
            texto_limpo += caractere

    if texto_limpo == texto_limpo[::-1]:
        print("A entrada é um palíndromo")
    else:
        print("Não é um palíndromo")

except ValueError as e:
    print(e)
```

### Exercício 3: Calculadora Simples com Tratamento de Erros

Objetivo: criar uma calculadora que trata `ValueError` (entrada inválida) e `ZeroDivisionError` (divisão por zero).

```python
print("Calculadora Simples")

try:
    numero1_info = input("Informe o primeiro número: ").strip()
    if numero1_info == "":
        raise ValueError("Erro: entrada vazia no primeiro número.")
    numero1 = float(numero1_info)

    numero2_info = input("Informe o segundo número: ").strip()
    if numero2_info == "":
        raise ValueError("Erro: entrada vazia no segundo número.")
    numero2 = float(numero2_info)

    operador = input("Informe o operador (+, -, *, /): ").strip()
    if operador not in ["+", "-", "*", "/"]:
        raise ValueError("Erro: informe um operador válido: +, -, * ou /.")

    if operador == "+":
        resultado = numero1 + numero2
    elif operador == "-":
        resultado = numero1 - numero2
    elif operador == "*":
        resultado = numero1 * numero2
    else:
        resultado = numero1 / numero2

    print(f"Resultado: {resultado}")

except ValueError as e:
    print(e)

except ZeroDivisionError:
    print("Erro: divisão por zero não é permitida.")
```
