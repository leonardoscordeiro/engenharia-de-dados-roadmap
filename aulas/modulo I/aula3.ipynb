### Exercício 1: Validador de Vendas

Você está analisando um conjunto de dados de vendas e precisa garantir que todos os registros tenham valores positivos para quantidade e preço. Escreva um programa que verifique esses campos e imprima "Dados válidos" se ambos forem positivos ou "Dados inválidos" caso contrário.

```python
print("Validador de vendas")

try:
    quantage = float(input("Informe a quantage: "))
    preco = float(input("Informe o preço: "))

    if quantage > 0 and preco > 0:
        print("Dados válidos")
    else:
        print("Dados inválidos")

except ValueError:
    print("Erro: informe apenas valores numéricos.")
```

### Exercício 4: Validação de Cadastro de Usuário

Antes de processar dados de usuários em um sistema de recomendação, garanta que cada usuário tenha idade entre 18 e 65 anos e tenha fornecido um email válido. Use `while` para repetir até que os dados estejam corretos.

```python

print ("Validador de e-mail")

print("Vamos realizar o seu cadastro")

while True:
    try:
        nome = input("Informe o seu nome: ").strip()

        if nome == "":
            raise ValueError("Nome inválido. Digite um nome válido.")

        print(f"Nome válido: {nome}")
        break

    except ValueError as e:
        print(e)

while True:
    try:
        age_info = input("Informe a sua age: ").strip()

        if age_info == "":
            raise ValueError("age inválida. Digite um valor válido.")

        age = int(age_info)

        if age < 18 or age > 65:
            raise ValueError("age inválida. Fora do range permitido.")

        print(f"age válida: {age}")
        break

    except ValueError as e:
        print(e)

while True:
    try:
        email = input("Informe o seu email: ").strip()

        if email == "":
            raise ValueError("Email inválido. Digite um email válido.")
        elif "@" not in email or "." not in email:
            raise ValueError("Email inválido. O email deve conter '@'.")

        print(f"Email válido: {email}")
        break

    except ValueError as e:
        print(e)


print ( "Informações registradas")
```

## Loop `for`

O loop `for` é utilizado para iterar sobre os itens de qualquer sequência (listas, strings, dicionários) e executar um bloco de código para cada item.

### Sintaxe básica

O `for` do Python itera sobre os itens de qualquer sequência (lista, string, etc.), na ordem em que aparecem. Diferente de C ou Pascal, não é necessário definir passo ou condição de parada manualmente.

```python
# Measure some strings:
words = ['cat', 'window', 'defenestrate']
for w in words:
    print(w, len(w))
```

```python
# Measure some strings:
nome = ['Luciano']
for letra in nome:
    print(letra)
```

### Usando `range()`

Se você precisa iterar sobre sequências numéricas, a função `range()` gera progressões aritméticas. O ponto de parada nunca é incluído.

```python
for i in range(5):
    print(i)
```

O ponto de parada fornecido nunca é incluído na lista; range(10) gera uma lista com 10 valores, exatamente os índices válidos para uma sequência de comprimento 10. É possível iniciar o intervalo com outro número, ou alterar a razão da progressão (inclusive com passo negativo):

```python
list(range(5, 10))
[5, 6, 7, 8, 9]

list(range(0, 10, 3))
[0, 3, 6, 9]

list(range(-10, -100, -30))
[-10, -40, -70]
```

### Iterando com `range()` e `len()`

Para acessar tanto o índice quanto o valor de uma sequência:

```python
a = ['Mary', 'had', 'a', 'little', 'lamb']
for i in range(len(a)):
    print(i, a[i])
```

### Exercício 6: Contagem de Palavras em Textos

Objetivo: dado um texto, contar quantas vezes cada palavra única aparece usando um dicionário como acumulador.

```python
texto = "a raposa marrom salta sobre a cadela preguiçosa"
palavras = texto.split()
contagem_palavras = {}

for palavra in palavras:
    if palavra in contagem_palavras:
        contagem_palavras[palavra] += 1
    else:
        contagem_palavras[palavra] = 1

print(contagem_palavras)
```

### Exercício 7: Normalização de Dados

Objetivo: normalizar uma lista de números para a escala de 0 a 1 usando a fórmula `(x - min) / (max - min)`.

```python
numeros = [10, 20, 30, 40, 50]
minimo = min(numeros)
maximo = max(numeros)
normalizados = [(x - minimo) / (maximo - minimo) for x in numeros]

print(normalizados)
```

### Exercício 8: Filtragem de Dados Faltantes

Objetivo: dada uma lista de dicionários representando usuários, filtrar aqueles que têm o campo `email` vazio usando list comprehension.

```python
usuarios = [
    {"nome": "Alice", "email": "alice@example.com"},
    {"nome": "Bob", "email": ""},
    {"nome": "Carol", "email": "carol@example.com"}
]

usuarios_validos = [usuario for usuario in usuarios if usuario["email"]]

print(usuarios_validos)
```

### Exercício 9: Extração de Subconjuntos

Objetivo: dada uma lista de números, extrair apenas os pares usando list comprehension.

```python
numeros = range(1, 11)
pares = [x for x in numeros if x % 2 == 0]

print(pares)
```

### Exercício 15: Processamento com Condição de Parada

Objetivo: processar itens de uma lista até encontrar um valor que indique parada (`"parar"`).

```python
itens = [1, 2, 3, "parar", 4, 5]

i = 0
while i < len(itens):
    if itens[i] == "parar":
        print("Parada encontrada, encerrando o processamento.")
        break
    # Processa o item
    print(f"Processando item: {itens[i]}")
    i += 1
```

### Exercício 16: Cadastro de Pessoas com Validação

Objetivo: integrar `while` com `try/except` para repetir o fluxo de cadastro até que o usuário insira dados válidos (nome, idade, cargo).

```python
print("Cadastro de Pessoas")


while True:
    try:
        name = input("Informe o seu nome: ").strip()

        if name == "":
            raise ValueError("Nome inválido. Digite um nome válido.")


        break

    except ValueError as e:
        print(e)


while True:
    try:
        age_info = input("Informe a sua idade: ").strip()

        if age_info == "":
            raise ValueError("Idade inválida. Digite uma idade válida.")

        age = int(age_info)

        if age <= 0:
            raise ValueError("Idade inválida. Digite um valor maior que zero.")

        break

    except ValueError as e:
        print(e)

        
while True:
    try:
        role = input("Informe o seu cargo: ").strip()

        if role == "":
            raise ValueError("Cargo inválido. Digite um nome válido.")


        break

    except ValueError as e:
        print(e)


print("========== CADASTRO ==========")
print (f"Nome: {name}")
print(f"Idade: {age}")
print (f"Cargo: {role}")

```
