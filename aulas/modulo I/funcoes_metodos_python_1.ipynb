## 1. `print()`
Mostra algo na tela.

```python
print("Olá")
```

## 2. `input()`
Recebe algo digitado pelo usuário.

```python
nome = input("Digite seu nome: ")
```

## 3. `type()`
Mostra o tipo da variável.

```python
idade = 26
print(type(idade))
```

## 4. `int()`
Converte para inteiro.

```python
numero = int("10")
print(numero)
```

## 5. `float()`
Converte para decimal.

```python
valor = float("10.5")
print(valor)
```

## 6. `str()`
Converte para texto.

```python
idade = str(26)
print(idade)
print(type(idade))
```

## 7. `len()`
Conta quantos itens tem.

```python
nomes = ["ana", "leo", "joao"]
print(len(nomes))
```

## 8. `sum()`
Soma valores.

```python
numeros = [1, 2, 3]
print(sum(numeros))
```

## 9. `max()`
Maior valor.

```python
numeros = [10, 50, 20]
print(max(numeros))
```

## 10. `min()`
Menor valor.

```python
numeros = [10, 50, 20]
print(min(numeros))
```

## 11. `sorted()`
Ordena e retorna nova lista.

```python
numeros = [5, 1, 9]
print(sorted(numeros))
```

## 12. `range()`
Gera sequência numérica.

```python
for i in range(5):
    print(i)
```

## 13. `enumerate()`
Percorre lista com índice e valor.

```python
nomes = ["ana", "leo", "joao"]
for indice, nome in enumerate(nomes):
    print(indice, nome)
```

## 14. `zip()`
Junta listas em pares.

```python
nomes = ["ana", "leo"]
idades = [20, 26]
for nome, idade in zip(nomes, idades):
    print(nome, idade)
```

## 15. `.strip()`
Remove espaços do começo e fim.

```python
nome = "  Leonardo  "
print(nome.strip())
```

## 16. `.lower()`
Deixa minúsculo.

```python
print("LEO".lower())
```

## 17. `.upper()`
Deixa maiúsculo.

```python
print("leo".upper())
```

## 18. `.split()`
Quebra texto em partes.

```python
frase = "maçã,banana,uva"
print(frase.split(","))
```

## 19. `.append()`
Adiciona item no final.

```python
nomes = ["ana", "leo"]
nomes.append("joao")
print(nomes)
```

## 20. `.remove()`
Remove pelo valor.

```python
nomes = ["ana", "leo", "joao"]
nomes.remove("leo")
print(nomes)
```

## Extras importantes em dicionários

### `.items()`
Percorre chave e valor do dicionário.

```python
dados = {"nome": "Leo", "idade": 26}
for chave, valor in dados.items():
    print(chave, valor)
```

### `.keys()`
Mostra as chaves.

```python
dados = {"nome": "Leo", "idade": 26}
print(dados.keys())
```

### `.values()`
Mostra os valores.

```python
dados = {"nome": "Leo", "idade": 26}
print(dados.values())
```

### `.get()`
Busca valor no dicionário com mais segurança.

```python
dados = {"nome": "Leo", "idade": 26}
print(dados.get("nome"))
```

## Exercício sugerido
Faça um programa que:
- peça nomes separados por vírgula
- use `.split()`
- use `.strip()`
- deixe tudo em minúsculo com `.lower()`
- coloque em uma lista
- mostre quantos nomes tem com `len()`

```python
entrada = input("Digite nomes separados por vírgula: ")

nomes = []
for nome in entrada.split(","):
    nomes.append(nome.strip().lower())

print(nomes)
print(len(nomes))
```
