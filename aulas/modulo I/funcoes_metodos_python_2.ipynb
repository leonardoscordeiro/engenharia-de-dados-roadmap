# Python intermediário — funções, estruturas e prática

Notebook com exemplos de nível intermediário, focado em raciocínio e repertório.

## 1. Funções com parâmetros e retorno

Uma função pode receber valores, processar e devolver um resultado com `return`.

```python
def calcular_media(notas):
    return sum(notas) / len(notas)

notas = [8, 7, 10, 9]
media = calcular_media(notas)

print(media)
```

## 2. Parâmetros com valor padrão

Você pode definir um valor padrão caso o usuário não envie nada.

```python
def saudacao(nome="amigo"):
    return f"Olá, {nome}"

print(saudacao())
print(saudacao("Leonardo"))
```

## 3. `return` x `print`

`print` mostra na tela. `return` devolve o valor para ser usado depois.

```python
def dobrar(numero):
    return numero * 2

resultado = dobrar(5)
print(resultado)
```

## 4. Condicionais dentro de função

Funções podem tomar decisões com `if`, `elif` e `else`.

```python
def classificar_temperatura(temperatura):
    if temperatura >= 26:
        return "Alta"
    elif temperatura <= 18:
        return "Baixa"
    else:
        return "Normal"

print(classificar_temperatura(30))
print(classificar_temperatura(20))
print(classificar_temperatura(15))
```

## 5. Laços `for` com funções

Você pode aplicar uma função em vários elementos.

```python
def quadrado(numero):
    return numero ** 2

numeros = [1, 2, 3, 4, 5]
quadrados = [quadrado(numero) for numero in numeros]

print(quadrados)
```

## 6. `while` para validação

Muito útil quando você quer repetir até o usuário acertar.

```python
while True:
    try:
        idade_info = input("Informe sua idade: ").strip()

        if idade_info == "":
            raise ValueError("Idade vazia.")

        idade = int(idade_info)

        if idade < 18 or idade > 65:
            raise ValueError("Idade fora do intervalo permitido.")

        print(f"Idade válida: {idade}")
        break

    except ValueError as e:
        print(e)
```

## 7. List comprehension

Forma enxuta de criar listas a partir de outra lista.

```python
numeros = [1, 2, 3, 4, 5]
dobrados = [numero * 2 for numero in numeros]

print(dobrados)
```

## 8. List comprehension com condição

```python
numeros = [1, 2, 3, 4, 5, 6]
pares = [numero for numero in numeros if numero % 2 == 0]

print(pares)
```

## 9. Dicionários

Guardam pares de chave e valor.

```python
pessoa = {
    "nome": "Leonardo",
    "idade": 26,
    "cidade": "Porto Alegre"
}

print(pessoa["nome"])
print(pessoa["idade"])
```

## 10. Percorrendo dicionários com `.items()`

```python
pessoa = {
    "nome": "Leonardo",
    "idade": 26,
    "cidade": "Porto Alegre"
}

for chave, valor in pessoa.items():
    print(f"{chave}: {valor}")
```

## 11. Contando ocorrências com dicionário

```python
texto = "banana"
contagem = {}

for letra in texto:
    if letra in contagem:
        contagem[letra] += 1
    else:
        contagem[letra] = 1

print(contagem)
```

## 12. Acumulando valores por categoria

```python
vendas = [
    {"categoria": "eletrônicos", "valor": 1200},
    {"categoria": "livros", "valor": 200},
    {"categoria": "eletrônicos", "valor": 800},
    {"categoria": "roupas", "valor": 350},
    {"categoria": "livros", "valor": 150}
]

total_por_categoria = {}

for venda in vendas:
    categoria = venda["categoria"]
    valor = venda["valor"]

    if categoria in total_por_categoria:
        total_por_categoria[categoria] += valor
    else:
        total_por_categoria[categoria] = valor

print(total_por_categoria)
```

## 13. `.get()` em dicionários

Ajuda a buscar chaves de forma mais segura.

```python
precos = {"maçã": 2.5, "banana": 3.0}

print(precos.get("maçã"))
print(precos.get("uva"))
print(precos.get("uva", 0))
```

## 14. Ordenação

`sorted()` retorna nova lista ordenada.

```python
numeros = [2, 360, 1000, 3, 14, 37, -1, 0]

ordenados = sorted(numeros)
print(ordenados)
```

## 15. Ordenando dicionários por valor

```python
produtos = {
    "banana": 4,
    "maca": 2,
    "pera": 7,
    "uva": 1
}

ordenados = sorted(produtos.items(), key=lambda item: item[1])

print(ordenados)
```

## 16. Tratamento de erros com `try-except`

Você pode tratar erros específicos.

```python
try:
    numero = int(input("Digite um número inteiro: "))
    print(numero)
except ValueError:
    print("Erro: informe um número inteiro válido.")
```

## 17. Função para normalizar nomes

```python
def normalizar_nome(nome):
    return nome.strip().lower()

nomes = [" Alice ", "BOB", "Carlos"]
nomes_normalizados = [normalizar_nome(nome) for nome in nomes]

print(nomes_normalizados)
```

## 18. Função para somar números de uma lista

```python
def somar_numeros(lista):
    return sum(lista)

numeros = [10, 20, 30, 40]
print(somar_numeros(numeros))
```

## 19. Função para validar email

Exemplo simples, sem regex.

```python
def validar_email(email):
    email = email.strip()

    if email == "":
        return False
    if "@" not in email or "." not in email:
        return False

    return True

print(validar_email("leo@gmail.com"))
print(validar_email("emailinvalido"))
```

## 20. Filtrando dados com condição

```python
produtos = [
    {"fruta": "banana", "quantidade": 2},
    {"fruta": "maca", "quantidade": 3},
    {"fruta": "banana", "quantidade": 5},
    {"fruta": "pera", "quantidade": 4}
]

for item in produtos:
    if item["fruta"] == "banana":
        print(item)
```

## 21. `enumerate()`

Muito útil quando você quer índice e valor.

```python
nomes = ["ana", "leo", "joao"]

for indice, nome in enumerate(nomes):
    print(indice, nome)
```

## 22. `zip()`

Junta listas em pares.

```python
nomes = ["ana", "leo", "joao"]
idades = [20, 26, 30]

for nome, idade in zip(nomes, idades):
    print(nome, idade)
```

## 23. `set()` para remover duplicados

```python
nomes = ["ana", "leo", "ana", "joao", "leo"]
nomes_unicos = list(set(nomes))

print(nomes_unicos)
```

## 24. Exercício guiado — total da compra

```python
fruits = ["maçã", "banana", "cereja"]
prices = {"maçã": 0.45, "banana": 0.30, "cereja": 0.65}

total_price = 0

for fruit in fruits:
    total_price += prices[fruit]

print(f"Preço total: R$ {total_price:.2f}")
```

## 25. Exercício guiado — palíndromo

```python
def eh_palindromo(texto):
    texto_limpo = ""

    for caractere in texto.lower():
        if caractere.isalnum():
            texto_limpo += caractere

    return texto_limpo == texto_limpo[::-1]

print(eh_palindromo("Ame a ema"))
print(eh_palindromo("Python"))
```

## 26. Exercício guiado — cadastro com validação

```python
while True:
    try:
        nome = input("Informe o nome: ").strip()

        if nome == "":
            raise ValueError("Nome inválido.")

        idade_info = input("Informe a idade: ").strip()

        if idade_info == "":
            raise ValueError("Idade vazia.")

        idade = int(idade_info)

        if idade < 18 or idade > 65:
            raise ValueError("Idade fora do intervalo permitido.")

        email = input("Informe o email: ").strip()

        if "@" not in email:
            raise ValueError("Email inválido.")

        print("Dados válidos")
        break

    except ValueError as e:
        print(e)
```
