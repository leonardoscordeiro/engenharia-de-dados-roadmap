## 1. Funções de alta ordem

Funções podem receber outras funções como argumento.

```python
def aplicar_operacao(lista, funcao):
    return [funcao(item) for item in lista]

def quadrado(x):
    return x ** 2

numeros = [1, 2, 3, 4]
resultado = aplicar_operacao(numeros, quadrado)

print(resultado)
```

## 2. `lambda`

Funções anônimas, úteis para operações curtas.

```python
numeros = [5, 1, 8, 3, 2]

ordenados = sorted(numeros, key=lambda x: x)
print(ordenados)

pares = list(filter(lambda x: x % 2 == 0, numeros))
print(pares)
```

## 3. `map()`, `filter()` e `reduce()`

```python
from functools import reduce

numeros = [1, 2, 3, 4, 5]

dobrados = list(map(lambda x: x * 2, numeros))
pares = list(filter(lambda x: x % 2 == 0, numeros))
soma_total = reduce(lambda acc, x: acc + x, numeros, 0)

print("dobrados:", dobrados)
print("pares:", pares)
print("soma_total:", soma_total)
```

## 4. Compreensões avançadas de lista

```python
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

achatada = [numero for linha in matriz for numero in linha]
print(achatada)

quadrados_pares = [x ** 2 for x in range(10) if x % 2 == 0]
print(quadrados_pares)
```

## 7. `*args` e `**kwargs`

Permitem receber quantidade variável de argumentos.

```python
def somar_todos(*args):
    return sum(args)

def mostrar_dados(**kwargs):
    for chave, valor in kwargs.items():
        print(f"{chave}: {valor}")

print(somar_todos(10, 20, 30, 40))

mostrar_dados(nome="Leonardo", idade=26, cidade="Porto Alegre")
```

## 11. Decorators

Permitem envolver comportamento de outra função.

```python
def log_execucao(func):
    def wrapper(*args, **kwargs):
        print(f"Executando: {func.__name__}")
        resultado = func(*args, **kwargs)
        print("Finalizado")
        return resultado
    return wrapper

@log_execucao
def saudacao(nome):
    return f"Olá, {nome}"

print(saudacao("Leonardo"))
```

## 12. Generators com `yield`

Geram valores sob demanda.

```python
def gerar_numeros(limite):
    for numero in range(limite):
        yield numero

for numero in gerar_numeros(5):
    print(numero)
```

## 14. Tratamento avançado de exceções

```python
def dividir(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Erro: divisão por zero."
    except TypeError:
        return "Erro: tipos inválidos."

print(dividir(10, 2))
print(dividir(10, 0))
print(dividir(10, "2"))
```

## 15. Levantando exceções personalizadas

```python
def validar_idade(idade):
    if idade < 0:
        raise ValueError("Idade não pode ser negativa.")
    if idade < 18:
        raise PermissionError("Acesso permitido apenas para maiores de idade.")
    return "Acesso permitido"

try:
    print(validar_idade(16))
except Exception as e:
    print(type(e).__name__, "-", e)
```

## 16. Classe e objetos

Python orientado a objetos na prática.

```python
class Produto:
    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco

    def aplicar_desconto(self, percentual):
        self.preco -= self.preco * (percentual / 100)

    def __repr__(self):
        return f"Produto(nome={self.nome!r}, preco={self.preco:.2f})"

produto = Produto("Notebook", 3500)
produto.aplicar_desconto(10)

print(produto)
```

## 17. Herança

```python
class Animal:
    def falar(self):
        return "Som genérico"

class Cachorro(Animal):
    def falar(self):
        return "Au au"

class Gato(Animal):
    def falar(self):
        return "Miau"

animais = [Cachorro(), Gato(), Animal()]

for animal in animais:
    print(animal.falar())
```

## 18. Métodos especiais (`dunder methods`)

```python
class Carrinho:
    def __init__(self):
        self.itens = []

    def adicionar(self, item):
        self.itens.append(item)

    def __len__(self):
        return len(self.itens)

    def __repr__(self):
        return f"Carrinho(itens={self.itens})"

carrinho = Carrinho()
carrinho.adicionar("banana")
carrinho.adicionar("maçã")

print(carrinho)
print(len(carrinho))
```

## 20. Agrupando dados manualmente

```python
vendas = [
    {"categoria": "eletrônicos", "valor": 1200},
    {"categoria": "livros", "valor": 200},
    {"categoria": "eletrônicos", "valor": 800},
    {"categoria": "roupas", "valor": 350},
    {"categoria": "livros", "valor": 150}
]

agrupado = {}

for venda in vendas:
    categoria = venda["categoria"]
    valor = venda["valor"]

    if categoria not in agrupado:
        agrupado[categoria] = []
    agrupado[categoria].append(valor)

print(agrupado)
```

## 21. `collections.Counter`

## 22. `defaultdict`

```python
from collections import defaultdict

vendas = [
    ("eletrônicos", 1200),
    ("livros", 200),
    ("eletrônicos", 800),
    ("livros", 150)
]

totais = defaultdict(int)

for categoria, valor in vendas:
    totais[categoria] += valor

print(dict(totais))
```

## 26. Função com validação robusta

```python
def converter_para_float_lista(texto):
    numeros = []

    for item in texto.split(","):
        item = item.strip()

        if item == "":
            raise ValueError("Item vazio encontrado na lista.")

        numeros.append(float(item))

    return numeros

try:
    entrada = "10, 20, 30.5, 40"
    lista = converter_para_float_lista(entrada)
    print(lista)
except ValueError as e:
    print(e)
```

## 27. Recursão

```python
def fatorial(n):
    if n == 0 or n == 1:
        return 1
    return n * fatorial(n - 1)

print(fatorial(5))
```

## 28. Memoization simples

```python
cache = {}

def fibonacci(n):
    if n in cache:
        return cache[n]
    if n <= 1:
        return n
    cache[n] = fibonacci(n - 1) + fibonacci(n - 2)
    return cache[n]

print(fibonacci(10))
print(cache)
```

## 29. `functools.lru_cache`

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n):
    if n <= 1:
        return n
    return fib(n - 1) + fib(n - 2)

print(fib(30))
```

## 33. Validação com regex

```python
import re

def validar_email(email):
    padrao = r"^[\w\.-]+@[\w\.-]+\.\w+$"
    return bool(re.match(padrao, email))

print(validar_email("leo@gmail.com"))
print(validar_email("email_invalido"))
```

## 34. Mini projeto — análise de vendas

```python
vendas = [
    {"produto": "Notebook", "categoria": "eletrônicos", "valor": 3500},
    {"produto": "Mouse", "categoria": "eletrônicos", "valor": 120},
    {"produto": "Livro A", "categoria": "livros", "valor": 80},
    {"produto": "Livro B", "categoria": "livros", "valor": 120},
    {"produto": "Camiseta", "categoria": "roupas", "valor": 90}
]

total_geral = sum(venda["valor"] for venda in vendas)

total_por_categoria = {}
for venda in vendas:
    categoria = venda["categoria"]
    valor = venda["valor"]
    total_por_categoria[categoria] = total_por_categoria.get(categoria, 0) + valor

produto_mais_caro = max(vendas, key=lambda item: item["valor"])

print("total_geral:", total_geral)
print("total_por_categoria:", total_por_categoria)
print("produto_mais_caro:", produto_mais_caro)
```
