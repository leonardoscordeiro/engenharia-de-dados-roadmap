## Definindo Funções

Para criar uma função em Python, use a keyword `def`, seguida de um nome descritivo, parênteses `()` com zero ou mais parâmetros, e dois pontos `:`. O bloco indentado é o corpo da função.

```python
def minha_funcao():
    return "Hello, World!"
```

## Nomes, Parâmetros e Argumentos

- **Nomes de funções** seguem as regras de variáveis: letras, números (não no início) e underscores. Use `snake_case` por convenção.
- **Parâmetros** são as variáveis na definição da função (placeholders).
- **Argumentos** são os valores reais passados na chamada.

```python
def soma(a, b):
    return a + b
```

## Palavras-chave Importantes

- `def` — inicia a definição de uma função
- `return` — retorna um valor da função (se omitido, retorna `None`)
- `pass` — placeholder para uma função vazia

## Chamando Funções

Use o nome da função seguido por parênteses com os argumentos necessários.

```python
resultado = soma(5, 3)
print(resultado)  # Saída: 8
```

## Valores Padrão e Argumentos Nomeados

Funções podem ter parâmetros com valores padrão, permitindo chamadas com menos argumentos.

```python
def cumprimentar(nome, mensagem="Olá"):
    print(f"{mensagem}, {nome}!")
```

Você também pode chamar funções com argumentos nomeados para maior clareza.

```python
cumprimentar(mensagem="Bem-vindo", nome="João")
```

## Funções com Type Hints

Vamos evoluir as funções adicionando type hints para tornar o código mais legível e seguro.

### 1. Calcular Média de Valores em uma Lista

```python
from typing import List

def calcular_media(valores: List[float]) -> float:
    return sum(valores) / len(valores)
```

### 2. Filtrar Dados Acima de um Limite

```python
def filtrar_acima_de(valores: List[float], limite: float) -> List[float]:
    resultado = []
    for valor in valores:
        if valor > limite:
            resultado.append(valor)
    return resultado
```

### 3. Contar Valores Únicos em uma Lista

```python
def contar_valores_unicos(lista: List[int]) -> int:
    return len(set(lista))
```

### 4. Converter Celsius para Fahrenheit em uma Lista

```python
def celsius_para_fahrenheit(temperaturas_celsius: List[float]) -> List[float]:
    return [(9/5) * temp + 32 for temp in temperaturas_celsius]
```

### 5. Calcular Desvio Padrão de uma Lista

```python
def calcular_desvio_padrao(valores: List[float]) -> float:
    media = sum(valores) / len(valores)
    variancia = sum((x - media) ** 2 for x in valores) / len(valores)
    return variancia ** 0.5
```

### 6. Encontrar Valores Ausentes em uma Sequência

```python
def encontrar_valores_ausentes(sequencia: List[int]) -> List[int]:
    completo = set(range(min(sequencia), max(sequencia) + 1))
    return list(completo - set(sequencia))
```

## Desafio: Processamento de Vendas com CSV

Aplique tudo o que aprendeu sobre funções para ler, processar e agregar dados de um arquivo CSV real.

### Tarefas

1. **Ler o arquivo CSV** e carregar os dados
2. **Processar os dados** em um dicionário agrupado por categoria
3. **Calcular o total de vendas** (Quantidade x Venda) por categoria

### Funções a implementar

| Função | Entrada | Saída |
|--------|---------|-------|
| `ler_csv` | Nome do arquivo CSV | Lista de dicionários |
| `processar_dados` | Lista de dicionários | Dicionário agrupado por categoria |
| `calcular_vendas_categoria` | Dicionário processado | Dicionário com total de vendas por categoria |

```python
import pandas as pd

vendas = pd.read_csv("vendas.csv")

print(vendas)
```

```python
# 1. Ler o CSV e devolver uma lista de dicionários
import csv

def ler_csv(nome_arquivo: str) -> list[dict]:
    with open(nome_arquivo, encoding="utf-8") as arquivo:
        leitor = csv.DictReader(arquivo)   # cada linha vira um dict
        dados = list(leitor)               # materializa tudo numa lista
    return dados
    
```

```python
# 2. Agrupar os produtos por categoria
def processar_dados(dados: list[dict]) -> dict:

    agrupado = {}
 
    for linha in dados:
        categoria = linha["Categoria"]
        produto = {
            "Produto":    linha["Produto"],
            "Categoria":  linha["Categoria"],
            "Quantidade": int(linha["Quantidade"]),
            "Venda":      float(linha["Venda"]),
        }
 
     
        if categoria not in agrupado:
            agrupado[categoria] = []
 
        agrupado[categoria].append(produto)
 
    return agrupado
```

```python
def calcular_vendas_categoria(agrupado: dict) -> dict:
    
    totais = {}
 
    for categoria, produtos in agrupado.items():
        total = 0
        for p in produtos:
            total += p["Quantidade"] * p["Venda"]
        totais[categoria] = total
 
    return totais
```

```python
dados = ler_csv("vendas.csv")
print("=== Dados brutos (lista de dicts) ===")
for d in dados:
    print(d)
```

```python
agrupado = processar_dados(dados)
print("\n=== Dados agrupados por categoria ===")
for cat, produtos in agrupado.items():
    print(f"\n{cat}:")
    for p in produtos:
        print(f"  {p}")
```

```python
totais = calcular_vendas_categoria(agrupado)
print("\n=== Total de vendas por categoria ===")
for cat, total in totais.items():
    print(f"  {cat}: R$ {total:,.2f}")
```
