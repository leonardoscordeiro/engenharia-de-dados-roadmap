### Exercício 3: Crie uma função chamada soma_lista que receba uma lista de números e retorne a soma de todos eles. Mas sem usar a função sum() — a ideia é você fazer a lógica na mão.

```python
def soma_lista(lista):
    soma = 0
    for item in lista:
        soma += item

    return soma 

print(soma_lista([1, 2, 3, 4]))    # → 10
print(soma_lista([10, 20, 30]))    # → 60
print(soma_lista([]))              # → 0
        
```

### Exercício 4: Crie uma função chamada contar_pares que receba uma lista de números e retorne quantos deles são pares.

```python
def contar_pares(numeros):
    pares = 0
    for numero in numeros:
        if numero % 2 == 0:
            pares += 1
            
    return pares

print(contar_pares([1, 2, 3, 4, 5, 6]))  # → 3
print(contar_pares([7, 9, 11]))           # → 0
print(contar_pares([2, 4, 6]))            # → 3
```

### Exercício 5: Crie uma função chamada inverter_texto que receba uma string e retorne ela invertida.

```python
def inverter_texto(texto):
    return texto[::-1]

print(inverter_texto("python"))  # → nohtyp
print(inverter_texto("ola"))     # → alo
```

 ### Exercício 6: Crie uma função chamada sem_duplicatas que receba uma lista e retorne uma nova lista sem elementos repetidos, mantendo a ordem original.

```python
def sem_duplicatas(duplicata):
    nova_lista = []
    for item in duplicata:
        if item not in nova_lista:
            nova_lista.append(item)
    return nova_lista


print(sem_duplicatas([1, 2, 2, 3, 1, 4]))      # → [1, 2, 3, 4]
print(sem_duplicatas(["a", "b", "a", "c"]))     # → ['a', 'b', 'c']
print(sem_duplicatas([5, 5, 5]))                 # → [5]
```

### Exercício 7: Crie uma função chamada resumo_lista que receba uma lista de números e retorne um dicionário com as seguintes informações: soma, média, maior e menor valor.

```python
def resumo_lista(lista):
    # soma
    soma = 0
    for item in lista:
        soma += item

    # média
    media = soma / len(lista)

    # maior
    maior = lista[0]
    for item in lista:
        if item > maior:
            maior = item

    # menor
    menor = lista[0]
    for item in lista:
        if item < menor:
            menor = item

    return {"soma": soma, "media": media, "maior": maior, "menor": menor}
```

###  Exercício 8: Crie uma função chamada validar_senha que receba uma string e retorne True se a senha for válida e False se não for. As regras são:
* Ter pelo menos 8 caracteres
* Conter pelo menos um número

```python
def validar_senha(senha):
    if len(senha) < 8:
        return False
    tem_numero = any(caractere.isdigit() for caractere in senha)
    
    if not tem_numero:
        return False
    else:
        return True
    
print(validar_senha("python2026"))  # → True
print(validar_senha("abcdefgh"))    # → False (sem número)
print(validar_senha("abc123"))      # → False (menos de 8 caracteres)
print(validar_senha("segura99"))    # → True
```

## Exercício 9: Crie uma função chamada contar_palavras que receba um texto e retorne um dicionário onde cada chave é uma palavra e o valor é quantas vezes ela aparece.

```python
def conta_palavras (texto):
    contagem = {}
    for palavra in texto.split():     
        if palavra in contagem:
            contagem[palavra] += 1
        else:
             contagem[palavra] = 1
    
    return contagem

print(conta_palavras("o gato viu o rato e o rato viu o gato"))
```

### Exercício 10: Crie uma função chamada calculadora que receba dois números e uma operação (string) e retorne o resultado.

```python
def calculadora(operacao):
    partes = operacao.split()
    a = int(partes[0])          # converte pra número
    b = int(partes[1])          # converte pra número
    operador = partes[2]

    if operador == "soma":
        return a + b
    elif operador == "subtracao":
        return a - b
    elif operador == "multiplicacao":
        return a * b
    elif operador == "divisao":
        if b == 0:
            return "Erro: divisão por zero"
        return a / b

print(calculadora("10 5 soma"))          # → 15
print(calculadora("10 5 subtracao"))     # → 5
print(calculadora("10 5 multiplicacao")) # → 50
print(calculadora("10 0 divisao"))       # → Erro: divisão por zero
```

### Exercício 11: Crie uma função chamada filtrar que receba uma lista de números e um valor mínimo, e retorne uma nova lista apenas com os números maiores ou iguais ao mínimo.

```python
def filtrar(lista,comparador) :
    maior_igual = []

    for numero in lista:
        if numero >= comparador:
            maior_igual.append(numero)

    return maior_igual

print(filtrar([1, 15, 8, 20, 3, 12], 10))     # → [15, 20, 12]
print(filtrar([5, 2, 9, 1], 3))                # → [5, 9]
print(filtrar([100, 200, 300], 500))           # → []    
```

### Exercício 12: Crie uma função chamada agrupar_por_paridade que receba uma lista de números e retorne um dicionário com duas chaves: "pares" e "impares", cada uma com a lista dos respectivos números.

```python
def agrupar_por_paridade(lista):

    dicionario = {"par":[], "impar":[]}

    for numero in lista:
        if numero % 2 == 0:
            dicionario["par"].append(numero)
        else:
            dicionario["impar"].append(numero)

    return dicionario    



print(agrupar_por_paridade([1, 2, 3, 4, 5, 6]))
# → {'pares': [2, 4, 6], 'impares': [1, 3, 5]}

print(agrupar_por_paridade([10, 15, 20]))
# → {'pares': [10, 20], 'impares': [15]}  
```

### Exercício 13: Crie uma função chamada aplicar que receba uma lista de números e uma função como parâmetros, e retorne uma nova lista com a função aplicada a cada item.

```python
def dobro(x):
    return x * 2

def quadrado(x):
    return x ** 2

def aplicar(numeros, funcao):
    resultado = []
    for numero in numeros:
        resultado.append(funcao(numero))   # chama a função recebida
    return resultado


    
print(aplicar([1, 2, 3, 4], dobro))     # → [2, 4, 6, 8]
print(aplicar([1, 2, 3, 4], quadrado))   # → [1, 4, 9, 16]
    
```

### Exercício 14: Crie uma função chamada buscar_aluno que receba uma lista de dicionários (alunos) e um nome, e retorne o dicionário do aluno encontrado. Se não encontrar, retorne "Aluno não encontrado".

```python
alunos = [
    {"nome": "Ana", "nota": 8.5},
    {"nome": "João", "nota": 6.0},
    {"nome": "Maria", "nota": 9.2}
    ]

def buscar_aluno (alunos, nome):
    
    for aluno in alunos:
        if aluno["nome"] == nome:
            return aluno                     
    return "Aluno não encontrado"            


print(buscar_aluno(alunos, "João"))    # → {'nome': 'João', 'nota': 6.0}
print(buscar_aluno(alunos, "Pedro"))   # → Aluno não encontrado
```

### Exercício 15: Crie uma função chamada adivinhar_numero que gere um número aleatório entre 1 e 10, e peça pro usuário adivinhar. O programa deve continuar pedindo até acertar, e no final dizer quantas tentativas foram necessárias.

```python
from random import randint

def adivinhar_numero():                          
    contador = 0
    numero_gerado = randint(1, 10)
    
    while True:
        try:
            numero_informado_info = input("Chute um número entre 1 e 10: ").strip()
            if numero_informado_info == "":
                raise ValueError("Erro. Valor inválido!")
            
            numero_informado = int(numero_informado_info)
            contador += 1                         

            if numero_informado == numero_gerado:
                print(f"Acertou em {contador} tentativas!")
                break
            else:
                print("Errou! Tente novamente.")

        except ValueError as e:
            print(e)

adivinhar_numero()
```

### Exercício 16: Crie uma função chamada notas_alunos que receba uma lista de dicionários (alunos com nome e nota) e retorne um novo dicionário classificando cada aluno como "aprovado" (nota >= 7) ou "reprovado" (nota < 7).

```python
alunos = [
    {"nome": "Ana", "nota": 8.5},
    {"nome": "João", "nota": 5.0},
    {"nome": "Maria", "nota": 9.2},
    {"nome": "Pedro", "nota": 6.5}
]


def notas_alunos (alunos):
    status = {"aprovado" : [],
              "reprovado" : []}
    
    for aluno in alunos:
        if aluno["nota"] >= 7.0:
            status["aprovado"].append(aluno["nome"])
        else:
            status["reprovado"].append(aluno["nome"])
    
    return(status)

print(notas_alunos(alunos))
```

### Exercício 17: Crie duas funções. A primeira, calcular_media, recebe uma lista de notas e retorna a média. A segunda, situacao_aluno, recebe o nome e uma lista de notas, usa calcular_media pra descobrir a média, e retorna se o aluno foi aprovado ou reprovado.

```python
def calcular_media(lista):
    media = sum(lista)/ len(lista)

    return media


def situacao_aluno (aluno, notas):
    
    media  = calcular_media(notas)

    if media >= 7.0:
        return f"{aluno}: aprovado com média {round(media, 2)}"
    else:
        return f"{aluno}: reprovado com média {round(media, 2)}"
    

print(situacao_aluno("Ana", [8.0, 7.5, 9.0]))
# → 'Ana: aprovada com média 8.17'

print(situacao_aluno("João", [5.0, 6.0, 4.5]))
# → 'João: reprovado com média 5.17'
```

### Exercício 18: Crie uma função chamada tupla_para_dicionario que receba uma lista de tuplas (pares chave-valor) e retorne um dicionário.

```python
pares = [("nome", "Ana"), ("idade", 25), ("cidade", "Porto Alegre")]

def tupla_para_dicionario(tupla):
    dicionario = {}
    for chave, valor in tupla:
        dicionario[chave] = valor
    return dicionario

print(tupla_para_dicionario(pares))
```
