 ### Exercício 6: Crie uma função chamada sem_duplicatas que receba uma lista e retorne uma nova lista sem elementos repetidos, mantendo a ordem original.

```python
def sem_duplicatas(duplicata):
    nova_lista = []
    for item in duplicata:
        if item not in nova_lista:
            nova_lista.append(item)
    return nova_lista


print(sem_duplicatas([1, 2, 2, 3, 1, 4]))      # → [1, 2, 3, 4]
print(sem_duplicatas(["a", "b", "a", "c"]))     # → ['a', 'b', 'c']
print(sem_duplicatas([5, 5, 5]))                 # → [5]
```

### Exercício 7: Crie uma função chamada resumo_lista que receba uma lista de números e retorne um dicionário com as seguintes informações: soma, média, maior e menor valor.

```python
def resumo_lista(lista):
    # soma
    soma = 0
    for item in lista:
        soma += item

    # média
    media = soma / len(lista)

    # maior
    maior = lista[0]
    for item in lista:
        if item > maior:
            maior = item

    # menor
    menor = lista[0]
    for item in lista:
        if item < menor:
            menor = item

    return {"soma": soma, "media": media, "maior": maior, "menor": menor}
```

* Conter pelo menos um número

### Exercício 10: Crie uma função chamada calculadora que receba dois números e uma operação (string) e retorne o resultado.

```python
def calculadora(operacao):
    partes = operacao.split()
    a = int(partes[0])          # converte pra número
    b = int(partes[1])          # converte pra número
    operador = partes[2]

    if operador == "soma":
        return a + b
    elif operador == "subtracao":
        return a - b
    elif operador == "multiplicacao":
        return a * b
    elif operador == "divisao":
        if b == 0:
            return "Erro: divisão por zero"
        return a / b

print(calculadora("10 5 soma"))          # → 15
print(calculadora("10 5 subtracao"))     # → 5
print(calculadora("10 5 multiplicacao")) # → 50
print(calculadora("10 0 divisao"))       # → Erro: divisão por zero
```

### Exercício 12: Crie uma função chamada agrupar_por_paridade que receba uma lista de números e retorne um dicionário com duas chaves: "pares" e "impares", cada uma com a lista dos respectivos números.

```python
def agrupar_por_paridade(lista):

    dicionario = {"par":[], "impar":[]}

    for numero in lista:
        if numero % 2 == 0:
            dicionario["par"].append(numero)
        else:
            dicionario["impar"].append(numero)

    return dicionario    



print(agrupar_por_paridade([1, 2, 3, 4, 5, 6]))
# → {'pares': [2, 4, 6], 'impares': [1, 3, 5]}

print(agrupar_por_paridade([10, 15, 20]))
# → {'pares': [10, 20], 'impares': [15]}  
```

```python
from random import randint

def adivinhar_numero():                          
    contador = 0
    numero_gerado = randint(1, 10)
    
    while True:
        try:
            numero_informado_info = input("Chute um número entre 1 e 10: ").strip()
            if numero_informado_info == "":
                raise ValueError("Erro. Valor inválido!")
            
            numero_informado = int(numero_informado_info)
            contador += 1                         

            if numero_informado == numero_gerado:
                print(f"Acertou em {contador} tentativas!")
                break
            else:
                print("Errou! Tente novamente.")

        except ValueError as e:
            print(e)

adivinhar_numero()
```

### Exercício 21: Crie uma função chamada relatorio_vendas que receba uma lista de dicionários (vendas) e retorne um dicionário com o resumo: total vendido, quantidade de vendas, ticket médio, e o vendedor que mais vendeu.

```python
vendas = [
    {"vendedor": "Ana", "valor": 150.0},
    {"vendedor": "João", "valor": 300.0},
    {"vendedor": "Ana", "valor": 200.0},
    {"vendedor": "Maria", "valor": 450.0},
    {"vendedor": "João", "valor": 100.0},
]


def relatorio_vendas(vendas):
    relatorio = {"total": [],
                "quantidade": [],
                "ticket_medio": [],
                "top_vendedor": []
                }
    
    
    relatorio["total"] = sum(venda["valor"] for venda in vendas)
    relatorio["quantidade"] = len(vendas)
    relatorio["ticket_medio"] = sum(venda["valor"] for venda in vendas) / len(vendas)
       
    
    vendas_por_vendedor = {}
    for venda in vendas:
        nome = venda["vendedor"]
        vendas_por_vendedor[nome] = vendas_por_vendedor.get(nome, 0) + venda["valor"]
    
    relatorio["top_vendedor"] = max(vendas_por_vendedor, key=vendas_por_vendedor.get)
    
    return relatorio

print(relatorio_vendas(vendas))
```

### Exercício 22: Crie uma função chamada processar_pedidos que receba uma lista de pedidos e retorne um dicionário com: total por status (entregue/pendente/cancelado), o pedido de maior valor, e a lista de clientes com pedidos pendentes.

```python
pedidos = [
    {"cliente": "Ana", "valor": 250.0, "status": "entregue"},
    {"cliente": "João", "valor": 180.0, "status": "pendente"},
    {"cliente": "Maria", "valor": 500.0, "status": "entregue"},
    {"cliente": "Pedro", "valor": 90.0, "status": "cancelado"},
    {"cliente": "Ana", "valor": 320.0, "status": "pendente"},
    {"cliente": "Lucas", "valor": 410.0, "status": "entregue"},
]


def processar_pedidos(pedidos):
    
    status = { 
        'total_por_status': {'entregue': 0.0, 'pendente': 0.0, 'cancelado': 0.0},
        'maior_pedido': {'cliente': '', 'valor': 0.0, 'status': ''},
        'clientes_pendentes': []
        }

    
    for pedido in pedidos:
       
        status_atual = pedido["status"]
        status["total_por_status"][status_atual] += pedido["valor"]

        if pedido["valor"] > status["maior_pedido"]["valor"]:
            status["maior_pedido"] = pedido 

        if status_atual == "pendente" and pedido["cliente"] not in status["clientes_pendentes"]:
            status["clientes_pendentes"].append(pedido["cliente"])

    return(status)

print(processar_pedidos(pedidos))
```

### Exercício 23: Crie uma função chamada analisar_logs que receba uma lista de dicionários (logs de sistema) e retorne: quantidade de erros por tipo, o horário do último erro crítico, e a lista de mensagens únicas.

```python
logs = [
    {"tipo": "INFO", "hora": "08:00", "mensagem": "Sistema iniciado"},
    {"tipo": "ERRO", "hora": "08:15", "mensagem": "Falha na conexão"},
    {"tipo": "ERRO", "hora": "09:30", "mensagem": "Timeout no banco"},
    {"tipo": "CRITICO", "hora": "10:00", "mensagem": "Servidor caiu"},
    {"tipo": "INFO", "hora": "10:05", "mensagem": "Sistema reiniciado"},
    {"tipo": "ERRO", "hora": "11:20", "mensagem": "Falha na conexão"},
    {"tipo": "CRITICO", "hora": "14:45", "mensagem": "Servidor caiu"},
]


def analisar_logs(logs):

    info_log = {
        'contagem_por_tipo': {'INFO': 0, 'ERRO': 0, 'CRITICO': 0},
        'ultimo_critico': '',
        'mensagens_unicas': []}
    
    for log in logs:

        tipo_atual = log["tipo"]

        info_log["contagem_por_tipo"][tipo_atual] += 1


        if tipo_atual == "CRITICO":

            info_log["ultimo_critico"] = log["hora"]

        if log["mensagem"] not in info_log["mensagens_unicas"]:
            info_log["mensagens_unicas"].append(log["mensagem"])

    return(info_log)




print(analisar_logs(logs))
# → {
#     'contagem_por_tipo': {'INFO': 2, 'ERRO': 3, 'CRITICO': 2},
#     'ultimo_critico': '14:45',
#     'mensagens_unicas': ['Sistema iniciado', 'Falha na conexão', 'Timeout no banco', 'Servidor caiu', 'Sistema reiniciado']
# }
```

### Exercício 24: Crie uma função chamada dashboard_produtos que receba uma lista de vendas de produtos e retorne um relatório completo: faturamento por categoria, produto mais vendido (por quantidade), média de preço por categoria, e produtos com estoque baixo (< 10 unidades).

```python
vendas = [
    {"produto": "Mouse", "categoria": "Periféricos", "preco": 89.90, "qtd": 45, "estoque": 12},
    {"produto": "Teclado", "categoria": "Periféricos", "preco": 159.90, "qtd": 30, "estoque": 8},
    {"produto": "Monitor", "categoria": "Monitores", "preco": 1299.90, "qtd": 12, "estoque": 3},
    {"produto": "Headset", "categoria": "Áudio", "preco": 249.90, "qtd": 25, "estoque": 15},
    {"produto": "Caixa de Som", "categoria": "Áudio", "preco": 199.90, "qtd": 18, "estoque": 7},
    {"produto": "Webcam", "categoria": "Periféricos", "preco": 299.90, "qtd": 20, "estoque": 5},
]

def dashboard_produtos(vendas):
    relatorio = {
        'faturamento_por_categoria': {'Periféricos': 0.0, 'Monitores': 0.0, 'Áudio': 0.0},
        'mais_vendido': '',
        'media_preco_por_categoria': {'Periféricos': 0.0, 'Monitores': 0.0, 'Áudio': 0.0},
        'estoque_baixo': []
    }
    contagem_itens = {'Periféricos': 0, 'Monitores': 0, 'Áudio': 0}
    mais_vendido_qtd = 0

    
    for venda in vendas:
        cat = venda["categoria"]
        
        # 1. Faturamento
        relatorio["faturamento_por_categoria"][cat] += venda["preco"] * venda["qtd"]

        # 2. Preparar Média (apenas soma os preços aqui)
        relatorio["media_preco_por_categoria"][cat] += venda["preco"]
        contagem_itens[cat] += 1

        # 3. Mais vendido
        if venda["qtd"] > mais_vendido_qtd:
            mais_vendido_qtd = venda["qtd"]
            relatorio["mais_vendido"] = venda["produto"]

        # 4. Estoque baixo
        if venda["estoque"] < 10:
            relatorio["estoque_baixo"].append(venda["produto"])

    
    for cat in relatorio["faturamento_por_categoria"]:
       
        relatorio["faturamento_por_categoria"][cat] = round(relatorio["faturamento_por_categoria"][cat], 2)
        
        # Calcula a média real
        soma_precos = relatorio["media_preco_por_categoria"][cat]
        total_produtos = contagem_itens[cat]
        relatorio["media_preco_por_categoria"][cat] = round(soma_precos / total_produtos, 2)

    return relatorio




print(dashboard_produtos(vendas))
# → {
#     'faturamento_por_categoria': {
#         'Periféricos': 14840.5,
#         'Monitores': 15598.8,
#         'Áudio': 9845.7
#     },
#     'mais_vendido': 'Mouse',
#     'media_preco_por_categoria': {
#         'Periféricos': 183.23,
#         'Monitores': 1299.9,
#         'Áudio': 224.9
#     },
#     'estoque_baixo': ['Teclado', 'Monitor', 'Caixa de Som', 'Webcam']
# }
```

### Exercício 26 (último!): Crie uma função chamada cruzar_dados que receba duas listas de dicionários (funcionários e departamentos) e retorne uma lista combinada com o nome do departamento incluído.

```python
funcionarios = [
    {"nome": "Ana", "dept_id": 1, "salario": 5000.0},
    {"nome": "João", "dept_id": 2, "salario": 3500.0},
    {"nome": "Maria", "dept_id": 1, "salario": 7000.0},
    {"nome": "Pedro", "dept_id": 3, "salario": 4000.0},
]

departamentos = [
    {"id": 1, "nome": "Engenharia"},
    {"id": 2, "nome": "Marketing"},
    {"id": 3, "nome": "Vendas"},
]



def cruzar_dados(funcionarios, departamentos):
    resultado = []
    dept_dict = {dept["id"]: dept["nome"] for dept in departamentos}

    for func in funcionarios:
        dept_nome = dept_dict.get(func["dept_id"], "Desconecido")
        resultado.append({
            "nome": func["nome"],
            "departamento": dept_nome,
            "salario": func["salario"]
        })

    return resultado

print(cruzar_dados(funcionarios, departamentos))
# → [
#     {'nome': 'Ana', 'departamento': 'Engenharia', 'salario': 5000.0},
#     {'nome': 'João', 'departamento': 'Marketing', 'salario': 3500.0},
#     {'nome': 'Maria', 'departamento': 'Engenharia', 'salario': 7000.0},
#     {'nome': 'Pedro', 'departamento': 'Vendas', 'salario': 4000.0}
# ]
```
