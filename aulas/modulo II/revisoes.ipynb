## Exercício 1: Relatório de Transações Bancárias

**Objetivo:** Criar uma função chamada `relatorio_transacoes` que receba uma lista de transações bancárias e retorne um relatório contendo:

| Item | Descrição |
|------|-----------|
| **Total por tipo** | Soma dos valores agrupados por `deposito`, `saque` e `transferencia` |
| **Maior transação** | A transação com o maior valor registrado |
| **Acima de R$ 1.000** | Quantidade de transações com valor ≥ 1000 |
| **Contas únicas** | Lista de todas as contas distintas |

> **Conceitos praticados:** funções, dicionários, loops `for`, condicionais, listas

### 1.1 Base de dados

Abaixo temos uma lista de dicionários representando transações bancárias. Cada transação possui três campos: `conta`, `tipo` e `valor`.

````python
transacoes = [
    {"conta": "001", "tipo": "deposito",      "valor": 1500.0},
    {"conta": "002", "tipo": "saque",          "valor": 300.0},
    {"conta": "001", "tipo": "transferencia",  "valor": 2000.0},
    {"conta": "003", "tipo": "deposito",       "valor": 800.0},
    {"conta": "002", "tipo": "deposito",       "valor": 1200.0},
    {"conta": "001", "tipo": "saque",          "valor": 500.0},
    {"conta": "003", "tipo": "transferencia",  "valor": 3500.0},
]

print(f"Total de transações: {len(transacoes)}")
print(f"Exemplo de transação: {transacoes[0]}")
````

### 1.2 Solução

**Estratégia passo a passo:**

1. Criar um dicionário `relatorio` com a estrutura do resultado esperado
2. Iterar sobre cada transação com `for`
3. A cada iteração:
   - Acumular o valor no tipo correspondente (`deposito`, `saque`, `transferencia`)
   - Comparar com a maior transação encontrada até agora
   - Verificar se o valor é ≥ 1000
   - Adicionar a conta à lista de contas únicas (se ainda não estiver lá)

````python
def relatorio_transacoes(transacoes):

    # Estrutura inicial do relatório
    relatorio = {
        "total_por_tipo": {"deposito": 0.0, "saque": 0.0, "transferencia": 0.0},
        "maior_transacao": {"conta": "", "tipo": "", "valor": 0.0},
        "acima_de_1000": 0,
        "contas_unicas": [],
    }

    for transacao in transacoes:

        # 1. Acumula o valor no tipo correspondente
        tipo_atual = transacao["tipo"]
        relatorio["total_por_tipo"][tipo_atual] += transacao["valor"]

        # 2. Verifica se é a maior transação até agora
        if transacao["valor"] >= relatorio["maior_transacao"]["valor"]:
            relatorio["maior_transacao"] = transacao

        # 3. Conta transações acima de R$ 1.000
        if transacao["valor"] >= 1000:
            relatorio["acima_de_1000"] += 1

        # 4. Adiciona conta se ainda não existir na lista
        if transacao["conta"] not in relatorio["contas_unicas"]:
            relatorio["contas_unicas"].append(transacao["conta"])

    return relatorio
````

### 1.3 Executando e conferindo o resultado

````python
resultado = relatorio_transacoes(transacoes)

print("=" * 50)
print("       RELATÓRIO DE TRANSAÇÕES BANCÁRIAS")
print("=" * 50)

print("\nTotal por tipo:")
for tipo, total in resultado["total_por_tipo"].items():
    print(f"  - {tipo.capitalize():20s} R$ {total:,.2f}")

maior = resultado["maior_transacao"]
print(f"\nMaior transação:")
print(f"  - Conta {maior['conta']} | {maior['tipo'].capitalize()} | R$ {maior['valor']:,.2f}")

print(f"\nTransações acima de R$ 1.000: {resultado['acima_de_1000']}")
print(f"Contas únicas: {resultado['contas_unicas']}")
````

